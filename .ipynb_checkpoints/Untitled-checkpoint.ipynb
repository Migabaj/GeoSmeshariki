```python
def find_place_index(json_obj, place):
    for i, feature in enumerate(json_obj['features']):
        if 'continent' in feature['properties'] and feature['properties']['continent'] == place:
            return i
        elif 'ADMIN' in feature['properties'] and feature['properties']['ADMIN'] == place:
            return i
```

```python
def set_properties(json_obj):
    for place in df['Place']:
        j_s = [(smeshariki_parts[df['Часть'][i]-1], stamp) for i, stamp in enumerate(df['Время']) if df['Place'][i] == place]
        if find_place_index(json_obj, place) or find_place_index(json_obj, place) == 0:
            json_obj['features'][find_place_index(json_obj, place)]['properties'] = {
                'name': place,
                'how_called': "\n".join([mesto for i, mesto in enumerate(df['Место']) if df['Place'][i] == place]),
                'episodes': "\n".join([episode for i, episode in enumerate(df['Серия']) if df['Place'][i] == place]),
                'quotes': "\n".join([f"{df['Персонаж'][i]}: {quote}" for i, quote in enumerate(df['Реплики']) if df['Place'][i] == place]),
                'links': "\n".join([f'https://www.youtube.com/watch?v={j[0]}&t={int(str(j[1]).split(":")[0])*60+int(str(j[1]).split(":")[1])}s' for j in j_s]),
            }
    return json_obj
```
